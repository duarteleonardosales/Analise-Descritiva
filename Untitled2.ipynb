```python
import numpy as np
from sklearn.preprocessing import normalize
import pandas as pd
import statsmodels.stats.outliers_influence

data = pd.read_csv('Dataset.csv', header='infer')

data[['ln(area)','bedroom','bathroom','garage']]
```

```python
data = data.to_numpy()
data = normalize(data, axis=0, norm='l2')

data_transposta = data.transpose()
x_x = np.dot(data_transposta, data)
k_index = np.sqrt(np.linalg.eigvals(x_x).max() / np.linalg.eigvals(x_x).min())

print(k_index)
```

- VIF - AREA: 2.71
- VIF - QUARTO: 2.06
- VIF - WC: 2.1
- VIF - VAGAS: 1.76

```python
test_mult_vif = []
for i in range(4):
  test_mult_vif.append(statsmodels.stats.outliers_influence.variance_inflation_factor(data,i+1))
print(test_mult_vif)
```
